```python
def mkdir(dir_name):
    try:
        if not(os.path.isdir(dir_name)):
            os.makedirs(os.path.join(dir_name))
    except OSError as e:
        if e.errno != errno.EEXIST:
            print("Failed to create directory!!!!!")
            raise
            
def video2frame(invideofilename, save_path, change=False):
    vidcap = cv2.VideoCapture(invideofilename)
    count = 0
    while True:
        success, image = vidcap.read()
        if not success:
            break
        if change:
            image = cv2.resize(image, (160, 120))
        print(success)
        print('Read a new frame: ', success)
        fname = "{}.jpg".format("{0:05d}".format(count))
        print(save_path + '_\\' + fname)
        if count %2 == 0:
            cv2.imwrite(save_path + '/origin/' + fname, image)
        else:
            cv2.imwrite(save_path + '/test/' + fname, image)
        count += 1
        if count == 30:
            break
        print("{} images are extracted in {}".format(count, save_path))
        
def video2frame_random(invideofilename, save_path, change=False):
    vidcap = cv2.VideoCapture(invideofilename)
    count = 0
    while True:
        success, image = vidcap.read()
        if not success:
            break
        if change:
            image = cv2.resize(image, (160, 120))
        print(success)
        print('Read a new frame: ', success)
        fname = "{}.jpg".format("{0:05d}".format(count))
        print(save_path + '_\\' + fname)
        cv2.imwrite(save_path + fname, image)
        count += 1
        if count == 15:
            break
        print("{} images are extracted in {}".format(count, save_path))
        
def video2frame_jump(invideofilename, save_path, change=False):
    vidcap = cv2.VideoCapture(invideofilename)
    count = 0
    count_1 = 0
    count_2 = 0
    while True:
        success, image = vidcap.read()
        if not success:
            break
        if change:
            image = cv2.resize(image, (160, 120))
        print(success)
        print('Read a new frame: ', success)
        fname = "{}.jpg".format("{0:05d}".format(count))
        print(save_path + '_\\' + fname)
        if count_1 < 15 and count % 2 == 0:
            cv2.imwrite(save_path + '/origin/' + fname, image)
            count_1 += 1
        elif count_2 < 15 and count % 3 == 0:
            cv2.imwrite(save_path + '/test/' + fname, image)
            count_2 += 1
        count += 1
        if count_1 == 15 and count_2 == 15:
            break
        if success == False:
            vidcap = cv2.VideoCapture(invideofilename)
        print("{} images are extracted in {}".format(count, save_path))
```
